```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
wtype = input('What weather type: ')
num_pts = input('How many data points: ')
```

```python
data = {'temp':'1','press':'3'}
url1='https://thingspeak.com/channels/1733232/fields/'
url2 = data[wtype]
url3 = '.csv?results='
url4 = str(num_pts)
#url = 'https://thingspeak.com/channels/1733232/fields/[wtype].csv?results=[num_pts]'
url = url1+url2+url3+url4
```

```python
plt.rcParams["figure.figsize"] = (20,5)
df = pd.read_csv(url)
```

```python
df.tail()
```

```python
plt.plot(df['field1'],color='#f68923')
```
